## Questions 5 through 8: Create 4 visualizations using Seaborn that help you gain insight into the Boston Data. Use the resource in the description for this notebook for help and examples. You can also use Seaborn documentation.

```python
# Question 5 Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap


df = pd.read_csv("/bostondata.csv")


df['status_dttm'] = pd.to_datetime(df['status_dttm'], errors='coerce')


sns.set(style="whitegrid")

# 1. Top 10 Most Common Violation Descriptions
plt.figure(figsize=(10, 6))
top_violations = df['description'].value_counts().head(10)
sns.barplot(y=top_violations.index, x=top_violations.values, palette="viridis")
plt.title("Top 10 Most Common Violation Descriptions")
plt.xlabel("Number of Cases")
plt.ylabel("Violation Description")
plt.tight_layout()
plt.show()
```

```python
# Question 6 Visualization
plt.figure(figsize=(6, 6))
status_counts = df['status'].value_counts()
sns.barplot(x=status_counts.index, y=status_counts.values, palette="Set2")
plt.title("Case Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```

```python
# Question 7 Visualization
plt.figure(figsize=(10, 6))
ward_counts = df['ward'].value_counts().sort_index()
sns.barplot(x=ward_counts.index, y=ward_counts.values, palette="coolwarm")
plt.title("Number of Violations by Ward")
plt.xlabel("Ward")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```

```python
# Question 8 Visualization
plt.figure(figsize=(12, 6))
monthly_violations = df.dropna(subset=['status_dttm']).copy()
monthly_violations['month'] = monthly_violations['status_dttm'].dt.to_period('M')
monthly_counts = monthly_violations['month'].value_counts().sort_index()
monthly_counts.plot(kind='line', marker='o', color='teal')
plt.title("Violations Over Time (by Month)")
plt.xlabel("Month")
plt.ylabel("Number of Violations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
